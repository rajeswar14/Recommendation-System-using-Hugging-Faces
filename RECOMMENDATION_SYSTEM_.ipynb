```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
```

```python
sentences = [" some great movie","another funny movie"]
encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors="pt")
encoded_input
```

we use **padding=True **

cause In machine learning, particularly when working with text data and neural networks, "padding = True" is used to ensure that all input sequences in a batch have the same length by adding extra "padding tokens" to shorter sequences, allowing the model to process data consistently without issues arising from varying input lengths; essentially, it prevents data loss at the edges by adding extra space around the actual data

we use **truncation = true cause**
True or 'longest_first' : truncate to a maximum length specified by the max_length argument or the maximum length accepted by the model if no max_length is provided ( max_length=None ). This will truncate token by token, removing a token from the longest sequence in the pair until the proper length is reached
make sure it token length  remains in the specified no of tokens it accepts
 ex:- bert models has max length of tokens as 520
   

```python
with torch.no_grad():
    model_output = model(**encoded_input)
model_output
# here we use pre trained models so no need of any gradient so we sue no grad here
# it has coverted our example sentences into tokens and that tokens  into embeddings
```

```python
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)
```

```python
sentence_embeddings = mean_pooling(model_output, encoded_input["attention_mask"])
sentence_embeddings
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
anime_embeddings = model.encode(df['description'].tolist())
```

```python
anime_embeddings
```

```python
query = "funny movies to watch"
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def get_recommendations(query, top_k=5):
  query_embedding = model.encode([query])
  similarities = cosine_similarity(query_embedding, anime_embeddings)
  top_indices = similarities[0].argsort()[-top_k:][::-1]
  return df.iloc[top_indices]
```

```python
recommendations = get_recommendations(query, 5)
recommendations
```
